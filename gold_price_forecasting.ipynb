```python
train_data =gold_price.head(699)
test_data  = gold_price.tail(32)
```

```python
train_data
```

```python
test_data
```

```python
#Linear Model
import statsmodels.formula.api as smf 
import numpy as np

linear_model = smf.ols('price~t',data=train_data).fit()

pred_linear =  linear_model.predict(test_data['t'])

rmse_linear = np.sqrt(np.mean((np.array(test_data['price'])-np.array(pred_linear))**2))
rmse_linear
```

```python
#Exponential

Exp = smf.ols('log_price~t',data=train_data).fit()

pred_Exp = Exp.predict(test_data['t'])

rmse_Exp = np.sqrt(np.mean((np.array(test_data['price'])-np.array(np.exp(pred_Exp)))**2))
rmse_Exp
```

```python
#Quadratic 

Quad = smf.ols('price~t+t_squared',data=train_data).fit()

pred_Quad = Quad.predict(test_data[["t","t_squared"]])

rmse_Quad = np.sqrt(np.mean((np.array(test_data['price'])-np.array(pred_Quad))**2))
rmse_Quad
```

```python
#Additive seasonality 

add_sea = smf.ols('price~Date',data=train_data).fit()

pred_add_sea = add_sea.predict(test_data[['Date']])

rmse_add_sea = np.sqrt(np.mean((np.array(test_data['price'])-np.array(pred_add_sea))**2))
rmse_add_sea
```

```python
#Additive Seasonality Quadratic 

add_sea_Quad = smf.ols('price~t+t_squared+month',data=train_data).fit()

pred_add_sea_quad = add_sea_Quad.predict(test_data[['month','t','t_squared']])

rmse_add_sea_quad = np.sqrt(np.mean((np.array(test_data['price'])-np.array(pred_add_sea_quad))**2))
rmse_add_sea_quad
```

```python
##Multiplicative Seasonality

Mul_sea = smf.ols('log_price~month',data = train_data).fit()

pred_Mult_sea = Mul_sea.predict(test_data[['month']])

rmse_Mult_sea = np.sqrt(np.mean((np.array(test_data['price'])-np.array(np.exp(pred_Mult_sea)))**2))

rmse_Mult_sea
```

```python
#Compare the results 

data = {"MODEL":pd.Series(["rmse_linear","rmse_Exp","rmse_Quad","rmse_add_sea","rmse_add_sea_quad","rmse_Mult_sea"]),
        "RMSE_Values":pd.Series([rmse_linear,rmse_Exp,rmse_Quad,rmse_add_sea,rmse_add_sea_quad,rmse_Mult_sea])}
table_rmse=pd.DataFrame(data)
table_rmse.sort_values(['RMSE_Values'])
```
